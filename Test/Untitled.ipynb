```python
# combine hive data with weather data
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import datetime

#wurzburg
# File to Load (Remember to change these)
ws_csv = "data/ws.csv"
ww_csv = "data/Wurzburg_Weather_Cleaned.csv"

# Read the hive Data
ws_df = pd.read_csv(ws_csv)
ww_df = pd.read_csv(ww_csv)

ws_df = ws_df.rename(columns={"day":"date"})
ww_df = ww_df.rename(columns={"T":"city_temp",
                             "U":"city_humid",
                             "Date":"date"})

ww_df.head()
```

```python
weather_hive_w_df = pd.merge(ws_df, ww_df, on="date", how="inner")
weather_hive_w_df.head()
```

```python
# Export file as a CSV, with the Pandas index and with the header
weather_hive_w_df.to_csv("weather_hive_w.csv", index=True, header=True)
```
